2. 시장 점유율 분석 (Pie Chart)
한 글로벌 IT 기업의 시장 점유율 데이터가 아래와 같이 주어졌습니다. 원 그래프를 이용하여 시장 점유율을 시각화하세요.
companies = ["A사", "B사", "C사", "D사", "기타"]
market_share = [35, 25, 20, 10, 10]
요구사항:
각 회사의 시장 점유율을 나타내는 원 그래프를 그리세요.
비율이 10% 이상인 경우 퍼센트를 표시하세요 (autopct="%1.1f%%")
그래프 제목을 "IT 기업 시장 점유율"로 설정하세요.
폭발 효과 (explode)를 사용하여 가장 높은 점유율을 가진 기업(A사)을 강조하세요.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 시스템에 설치된 한글 폰트 설정 (Windows 기준)
plt.rc('font', family='Malgun Gothic')  # '맑은 고딕' 폰트 사용
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 데이터
products = ["노트북", "태블릿", "스마트폰", "스마트워치", "헤드폰"]
sales = [150, 80, 300, 120, 180]

x = products
y = sales

plt.xlabel("제품명")
plt.ylabel("판매량")
plt.bar(x, y, color="green")
plt.title("전자제품 판매량 비교")
plt.show()
```

```python
companies = ["A사", "B사", "C사", "D사", "E사", "F사"]
market_share = np.array([40, 25, 15, 8, 7, 5])  # 일부 10% 미만 포함

plt.title("IT 기업 시장 점유율")
explode = (0.1, 0, 0, 0, 0, 0)  # A사 부분 돌출
plt.pie(market_share, labels=companies, autopct="%1.1f%%", explode=explode)
plt.show()


```

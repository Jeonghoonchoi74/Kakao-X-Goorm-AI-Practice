문제 3. 그래프 해석
함수 f(x)=2x+3 와
함수 f(x)= x^2 - 6x + 8 의 그래프에 대해 다음을 구하고, 맷플롯립으로 시각화하시오.
(1) 꼭짓점의 좌표
(2) x축과 만나는 점 (x절편)
(3) y축과 만나는 점 (y절편)
.

```python
import matplotlib.pyplot as plt
import numpy as np

# 시스템에 설치된 한글 폰트 설정 (Windows 기준)
plt.rc('font', family='Malgun Gothic')  # '맑은 고딕' 폰트 사용
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 꼭짓점의 좌표를 구하는 함수
def vertex(a, b, c):
    x = -b / (2 * a)
    y = c - (b**2) / (4 * a)
    return int(x), int(y)

# y절편을 구하는 함수
def y_intercept(a, b, c):
    result = c - (b**2) / (4 * a)
    return result

# x 절편의 음수를 구하는 함수
def negative_x_intercept(a, b, c):
    negative_x = (-b - np.sqrt(b**2 - 4 * a * c)) / (2 * a)
    return int(negative_x)

# x 절편의 양수를 구하는 함수
def positive_x_intercept(a, b, c):
    positive_x = (-b + np.sqrt(b**2 - 4 * a * c)) / (2 * a)
    return int(positive_x)

# 2차 함수
x_quadratic = np.linspace(-5, 5, 100)
y_quadratic1 = 2 * x_quadratic + 3
y_quadratic2 = x_quadratic**2 + -6 * x_quadratic + 8
plt.plot(x_quadratic, y_quadratic1, label='y = 2x+3')
plt.plot(x_quadratic, y_quadratic2, label='y = x^2 - 6x + 8')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2차 함수 그래프')
plt.legend()
plt.grid(True)
plt.show()

print("꼭짓점의 좌표", vertex(1, -6, 8))
print("y 절편 : ", y_intercept(1, -6, 8))
print(f"x 절편 : ({negative_x_intercept(1, -6, 8)},0) , ({positive_x_intercept(1, -6, 8)},0)")
```

## 배경 이야기
1. 오래된 성벽 너머로 붉은 노을이 드리워진 어느 밤, 숲속에 은은한 마법빛이 어른거리며 퍼지고 있었다.  
2. 그 빛 속에서 나타난 이는 십 년 전 행방불명된 왕자였다.  
3. 왕자는 어릴 적 공주와 함께 뛰놀던 성을 떠난 뒤, 마법의 비밀을 찾아 떠돌았다고 했다.  
4. 마을 사람들은 그가 곧 돌아오리라 믿었으나, 소식은 묘연했다.  
5. 그러던 중 사악한 마법사가 공주를 납치해 검은 탑에 가두었다는 소식이 들려왔다.  
6. 탑 주변에는 강력한 보호 룬이 깔려 있어, 지상(地上)에서는 결코 문을 열 수 없었다.  
7. 마법사는 공주의 용기를 꺾어 자신의 어둠의 제단에 바치려 했다.  
8. 왕자는 칼 대신 오래된 주문서를 펼치고, **‘용의 영혼 결정’**을 소환하는 의식을 준비했다.  
9. 이 의식은 용에게서 **심장의 일부**가 아니라, 용의 **영혼 일부**—즉 생명력을 나누어 받는 계약이다.  
10. 계약이 성립되면 용은 전신을 잃지 않고, 대신 **영혼의 수정 조각**을 하나 내어준다.  
11. 왕자는 그 수정 조각을 가슴에 품고 탑 앞으로 향했다.  
12. 의식을 시작하자 땅이 떨리고, 숲속 깊은 동굴에서 거대한 용 한 마리가 모습을 드러냈다.  
13. 용은 왕자의 결의와 용기를 시험하고자 깊은 눈빛으로 그를 내려다보았다.  
14. 왕자가 조용히 말하였다.   
15. 용은 고개를 끄덕이며, 날개 아래에서 빛나는 **영혼의 결정**을 한 조각 흩뿌렸다.  
16. 그 수정이 반짝이며 왕자에게 흡수되자, 그의 손끝에서 황금빛 마법 에너지가 맹렬히 타올랐다.  
17. 왕자는 에너지를 집중해 탑을 감싸던 검은 룬을 하나씩 해체했다.  
18. 마법진이 풀리자 탑 문이 삐걱이며 열렸다.  
19. 공주는 무사히 모습을 드러내었고, 왕자의 눈빛을 보며 안도의 한숨을 내쉬었다.  
20. 그러나 마법사는 마지막 남은 힘으로 탑을 무너뜨려 둘을 가두려 했다.  
21. 그때 동굴 위 하늘에서 용이 내지른 **거대한 포효(Protego Roar)**가 울려 퍼져, 무너지는 탑의 잔해를 막아냈다.  
22. 포효의 충격파가 마법사의 어둠의 마력을 사방으로 흩어뜨리자, 그는 힘을 잃고 땅에 쓰러졌다.  
23. 왕자와 공주는 용의 거대한 날개 아래에서 안전히 벗어났다.  
24. 마법사는 결국 동굴 아래 갇혀버렸고, 왕국에는 고요한 평화가 찾아왔다.  
25. 왕자와 공주는 성으로 돌아와, **용의 영혼 결정**을 성 안 신성 보관소에 안치했다.  
26. 이후 매년 마법 축제가 열려, 용과 맺은 약속을 기념하며 희망을 노래했다.  
27. 공주는 새롭게 얻은 용기와 지혜로 백성을 이끌었고,  
28. 왕국 사람들은 “진정한 힘은 희생이 아닌, 생명을 이어주는 계약에서 온다”는 교훈을 전해 들었다.  
29. 그리하여 용의 영혼 결정은 단순한 마법물이 아니라,  
30. **용기와 신뢰의 상징**으로 영원히 빛나게 되었다.

```python
import pandas as pd

df = pd.read_table('fairy_tale.txt')
df
```

```python
from konlpy.tag import Okt
import re

# 형태소 분석기 객체 생성
okt = Okt()

# 불용어 리스트(예시, 필요시 추가/수정)
stopwords = set([
    # 조사 및 접속 부사류
    '은', '는', '이', '가', '을', '를', '에', '의', '도', '로', '과', '와', '한', '에서', '이며', '에는', '에서는', '이다',
    '에게서', '으로 ',
    # 일반 불용어성 동사/형용사
    '했다', '하였다', '되었다', '있었다', '없었다', '된다', '되다', '하다', '있다', '없다', '이며', '되면', '되자', '하며', '하고자', '하자', '하려고', '한다', '된다',

    # 맥락성 약한 부사/조사 또는 중복명사
    '그리고', '그러나', '그러던', '그때', '이후', '이든', '어느', '어릴', '같은', '하는', '보며', '하여', '하며', '또한', '그리하여', '그러나', '그러던', '이것', '저것', '것', '수', '들', '그', '저',

    # 의미 약한 명사
    '모습', '눈빛', '손끝', '마리', '잔해', '고개', '결코', '퍼지고', '드러냈다', '보며', '안도', '한숨', '내쉬었다', '깊은', '떨리고', '떨어지다', '벗어나다', '보관', '안치', '맺은', '기념',

    # 빈도 낮은 의미 약한 표현 (후보)
    '되었다', '온다는', '이어주는', '빛나게', '흩어', '결국', '드러내었고', '나누어', '받는', '주는', '주는', '얻은', '새롭게', '내려다보았다', '끄덕', '조용히', '말하였다', '지른', '울려', '퍼져', '무너', '뜨려', '가두려', '막아', '냈다'
])

allow_single_chars = {'용', '성', '힘', '신', '탑'}
# 텍스트 전처리 및 토큰화 함수
def preprocess_and_tokenize(text):
    # 소문자 변환 (한글에는 영향 없음, 영문이 섞인 경우 대비)
    text = text.lower()
    # 특수문자, 숫자 제거
    text = re.sub(r'[^가-힣\s]', '', text)
    # 형태소 단위로 토큰화
    tokens = okt.morphs(text)
    # 길이 1 이하, 불용어 제거 
    tokens = [t for t in tokens if (len(t) > 1 or t in allow_single_chars) and t not in stopwords]
    return tokens

# 데이터프레임에서 텍스트 컬럼 추출
sentences = df.iloc[:, 0].tolist()
tokenized_sentences = [preprocess_and_tokenize(sent) for sent in sentences]

# 결과 확인
for i, tokens in enumerate(tokenized_sentences):
    print(f"{i+1}번째 문장 토큰: {tokens}")

# 단어 빈도수 계산
from collections import Counter
# 모든 문장에서 단어를 추출하여 빈도수 계산
all_tokens = [token for tokens in tokenized_sentences for token in tokens]
word_counts = Counter(all_tokens)
word_counts
```

```python
from gensim.models import Word2Vec

# CBOW 모델 학습
model_cbow = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=50,
    window=3,
    min_count=1,
    sg=0  # CBOW
)

```

```python
# Skip-gram 모델 학습
model_sg = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=50,
    window=3,
    min_count=1,
    sg=1  # Skip-gram
)
```

```python
# 각 모델별 시간 비교
import time
start_time = time.time()
model_cbow.train(tokenized_sentences, total_examples=len(tokenized_sentences), epochs=10)
end_time = time.time()
print(f"CBOW 모델 학습 시간: {end_time - start_time:.2f}초")
start_time = time.time()
model_sg.train(tokenized_sentences, total_examples=len(tokenized_sentences), epochs=10)
end_time = time.time()
print(f"Skip-gram 모델 학습 시간: {end_time - start_time:.2f}초")
```

```python
# CBOW 모델과 Skip-gram 모델에서 각 키워드별 유사 단어 상위 5개 출력
keywords = ['왕자', '공주', '마법', '용']

print("CBOW 모델 유사 단어")
for word in keywords:
    print(f"\n[{word}]")
    try:
        for similar_word, score in model_cbow.wv.most_similar(word, topn=5):
            print(f"{similar_word}: {score:.4f}")
    except KeyError:
        print("단어가 vocabulary에 없습니다.")

print("\nSkip-gram 모델 유사 단어")
for word in keywords:
    print(f"\n[{word}]")
    try:
        for similar_word, score in model_sg.wv.most_similar(word, topn=5):
            print(f"{similar_word}: {score:.4f}")
    except KeyError:
        print("단어가 vocabulary에 없습니다.")
```

```python
from sklearn.manifold import TSNE
import numpy as np

import matplotlib.pyplot as plt

# 한글 폰트 설정 (Windows 기준)
plt.rcParams['axes.unicode_minus'] = False


# 시각화에 사용할 단어 리스트 (주요 단어 + 상위 빈도 단어 일부)
highlight_words = ['왕자', '공주', '마법', '용', '성']
vocab = list(model_cbow.wv.index_to_key)
# 상위 50개 단어만 시각화 (너무 많으면 복잡)
words_to_plot = list(set(highlight_words + vocab[:100]))

# 각 단어의 임베딩 벡터 추출
word_vectors = np.array([model_cbow.wv[word] for word in words_to_plot])

# t-SNE로 2차원 축소
tsne = TSNE(n_components=2, random_state=42, perplexity=5)
word_vec_2d = tsne.fit_transform(word_vectors)

plt.figure(figsize=(10, 8))
for i, word in enumerate(words_to_plot):
    x, y = word_vec_2d[i]
    if word in highlight_words:
        plt.scatter(x, y, color='red', s=100, label=word if word not in plt.gca().get_legend_handles_labels()[1] else "")
        plt.text(x+0.1, y+0.1, word, fontsize=14, color='red', fontweight='bold')
    else:
        plt.scatter(x, y, color='gray', s=40)
        plt.text(x+0.1, y+0.1, word, fontsize=10, color='black')

plt.title('Word2Vec 임베딩(t-SNE) 시각화 (CBOW)')
plt.legend()
plt.show()
```

## 🔍 CBOW vs Skip-gram 결과 해석 및 비교

### ✅ 1. CBOW vs Skip-gram 비교

| 항목           | CBOW        | Skip-gram   |
| ------------ | ----------- | ----------- |
| **학습 속도**    | 빠름          | 느림          |
| **문맥 예측 방향** | 주변 → 중심 단어  | 중심 → 주변 단어  |
| **희귀 단어 학습** | 어려움         | 상대적으로 유리    |
| **데이터 크기**   | 데이터 많을 때 적합 | 데이터 적을 때 유리 |

---

### ✅ 2. 유사 단어 예시

* **`왕자`**

  * CBOW: `뿌리다`, `신성`, `삐걱`
  * Skip-gram: `거대하다`, `받다`, `신성`
* **`용`**

  * CBOW & Skip-gram 공통: `수정`, `붉다`, `잃다`

※ 둘 다 전반적으로 의미 유사성보다는 **문장 동시 등장 기반 유사도**를 반영

---

### ✅ 3. t-SNE 시각화 해석

* `왕자`는 `계약`, `결의`, `갇혀` 등과 가까이 위치 → 이야기 전개상 타당
* `마법`은 `탑`, `성`, `무능` 등과 연결 → 사건 장소 중심 클러스터링
* 핵심 키워드들이 의미 단위로 어느 정도 **군집(클러스터)** 형성

---

### ✅ 4. 동화 코퍼스 특성상 더 적합한 방식은?

* 코퍼스가 **짧고 희귀 단어 많음**
* Skip-gram은 희귀 중심 단어 학습에 유리
* 따라서 **Skip-gram이 더 적합**

---

### ✅ 결론

* **CBOW**는 빠르지만 작은 데이터에선 과학습 우려
* **Skip-gram**은 단어 수 적은 코퍼스에서 안정적
* **현재 동화 분석에는 Skip-gram 방식이 더 적합함**

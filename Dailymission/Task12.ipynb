```python
import pandas as pd

data = {
    "Title": [
        "Inception", "Interstellar", "The Dark Knight", "Parasite", "Avengers: Endgame",
        "The Matrix", "Gladiator", "Titanic", "Joker", "Mad Max: Fury Road",
        "The Godfather", "Pulp Fiction", "Schindler's List", "The Shawshank Redemption", "Fight Club"
    ],
    "Genre": [
        "Sci-Fi", "Sci-Fi", "Action", "Drama", "Action",
        "Sci-Fi", "Historical", "Romance", "Drama", "Action",
        "Crime", "Crime", "Drama", "Drama", "Drama"
    ],
    "Year": [
        2010, 2014, 2008, 2019, 2019,
        1999, 2000, 1997, 2019, 2015,
        1972, 1994, 1993, 1994, 1999
    ],
    "Director": [
        "Christopher Nolan", "Christopher Nolan", "Christopher Nolan", "Bong Joon-ho", "Anthony Russo",
        "Lana Wachowski", "Ridley Scott", "James Cameron", "Todd Phillips", "George Miller",
        "Francis Ford Coppola", "Quentin Tarantino", "Steven Spielberg", "Frank Darabont", "David Fincher"
    ],
    "Rating": [
        8.8, 8.6, 9.0, 8.6, 8.4,
        8.7, 8.5, 7.8, 8.5, 8.1,
        9.2, 8.9, 8.9, 9.3, 8.8
    ],
    "Audience_Million": [
        100, 120, 150, 80, 200,
        90, 50, 220, 85, 75,
        110, 95, 85, 130, 70
    ],
    "Budget_Million": [
        160, 165, 185, 11, 356,
        63, 103, 200, 55, 150,
        6, 8, 22, 25, 63
    ]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
"""
1. ÌäπÏ†ï Ïû•Î•¥Ïùò ÏòÅÌôî Î™©Î°ù Ï∂úÎ†•
Ïû•Î•¥Í∞Ä "Action"Ïù∏ ÏòÅÌôîÎßå Ï∂úÎ†•ÌïòÏÑ∏Ïöî.
üìå ÌûåÌä∏: df["Genre"] == "Action"ÏùÑ ÌôúÏö©ÌïòÏó¨ ÌïÑÌÑ∞ÎßÅ."
"""
genre = df[df["Genre"] == "Action"]

genre
```

```python
"""
2. ÌäπÏ†ï Ïó∞ÎèÑ Ïù¥ÌõÑ Í∞úÎ¥âÌïú ÏòÅÌôî Î™©Î°ù
2010ÎÖÑ Ïù¥ÌõÑ Í∞úÎ¥âÌïú ÏòÅÌôîÎßå Ï∂úÎ†•ÌïòÏÑ∏Ïöî.
"""
df[df["Year"] >= 2010]
```

```python
"""
3. ÌèâÏ†êÏù¥ 9.0 Ïù¥ÏÉÅÏù∏ ÏòÅÌôî Ï∞æÍ∏∞
ÌèâÏ†êÏù¥ 9.0 Ïù¥ÏÉÅÏù∏ ÏòÅÌôîÎßå Ï∂úÎ†•ÌïòÏÑ∏Ïöî.
"""
df[df["Rating"]>=9.0]
```

```python
"""
4. ÌäπÏ†ï Í∞êÎèÖÏùò ÏòÅÌôî Î™©Î°ù Ï∂úÎ†•
"Christopher Nolan"Ïù¥ Í∞êÎèÖÌïú ÏòÅÌôîÎßå Ï∂úÎ†•ÌïòÏÑ∏Ïöî.
"""
df[df["Director"] == "Christopher Nolan" ]
```

```python
"""
5. ÌäπÏ†ï ÏòàÏÇ∞ Î≤îÏúÑ ÎÇ¥Ïùò ÏòÅÌôî Ï∞æÍ∏∞
Ï†úÏûëÎπÑÍ∞Ä 50Î∞±Îßå ~ 200Î∞±Îßå ÏÇ¨Ïù¥Ïù∏ ÏòÅÌôîÎßå Ï∂úÎ†•ÌïòÏÑ∏Ïöî.
"""
df[(df["Budget_Million"] > 50) & (df["Budget_Million"] < 200)]
```

```python
"""
6. Ï†úÏûëÎπÑ 100Î∞±Îßå Ïù¥ÏÉÅÏù∏ ÏòÅÌôîÎßå ÌïÑÌÑ∞ÎßÅ
Ï†úÏûëÎπÑÍ∞Ä 100Î∞±Îßå Ïù¥ÏÉÅÏù∏ ÏòÅÌôî Î™©Î°ù Ï∂úÎ†•
"""
df[df["Budget_Million"]>=100]
```

```python
"""
7. Í¥ÄÍ∞ù ÏàòÍ∞Ä 100Î∞±Îßå Ïù¥ÏÉÅÏù∏ ÏòÅÌôîÎßå Ï∞æÍ∏∞
Í¥ÄÍ∞ù ÏàòÍ∞Ä 100Î∞±Îßå Ïù¥ÏÉÅÏù∏ ÏòÅÌôî Ï∂úÎ†•
"""

df[df["Audience_Million"]>=100]
```

```python
"""
8. Í∞ÄÏû• ÏµúÍ∑ºÏóê Í∞úÎ¥âÌïú ÏòÅÌôî Ï∞æÍ∏∞
Ïó∞ÎèÑÍ∞Ä Í∞ÄÏû• ÌÅ∞ Í∞íÏóê Ìï¥ÎãπÌïòÎäî ÏòÅÌôî Ï∞æÍ∏∞ (max() ÏÇ¨Ïö© Í∏àÏßÄ)
üìå ÌûåÌä∏: .iloc[]ÏùÑ ÌôúÏö©ÌïòÏó¨ Ï†ïÎ†¨ ÌõÑ ÎßàÏßÄÎßâ Ìñâ ÏÑ†ÌÉù.
"""
year = df.sort_values(by="Year", ascending=False)
max_year = year.iloc[0]["Year"]
max_year
df[df["Year"] == max_year]
```

```python
"""
9. Í∞ÄÏû• Ïò§ÎûòÎêú ÏòÅÌôî Ï∞æÍ∏∞
Ïó∞ÎèÑÍ∞Ä Í∞ÄÏû• ÏûëÏùÄ Í∞íÏóê Ìï¥ÎãπÌïòÎäî ÏòÅÌôî Ï∞æÍ∏∞ (min() ÏÇ¨Ïö© Í∏àÏßÄ)
"""
year = df.sort_values(by="Year", ascending=True)
min_year = year.iloc[0]["Year"]
min_year
df[df["Year"] == min_year]
```

## 문제 6. k 값의 영향
KNN 알고리즘에서 **k 값을 너무 작게** 설정하면 생길 수 있는 현상은?

1. 과소적합  
2. 계산량 증가  
3. 이상치에 민감해짐  
4. 모델이 무작위로 예측함  

👉 정답: ( 3 )  
👉 간단한 이유: 예를 들어 k=1일 경우, 예측 대상은 가장 가까운 하나의 데이터만 보고 판단하게 된다다. 이 이웃이 이상치 라면, 이를 정상적인 데이터로 잘못 인식하고 예측 결과를 왜곡시킬 수 있다. 
    이후 유사한 위치에 있는 새로운 데이터들도 이상치와 같은 클래스로 분류될 수 있기 때문에, k값이 너무 작으면 이상치에 민감해져 모델의 일반화 성능이 떨어질 수 있다.

## 문제 7. KNN과 데이터 스케일링
두 특성(feature)을 가진 데이터셋에서, 하나는 0~1 범위의 값(예: 키), 다른 하나는 0~10,000 범위의 값(예: 연봉)일 때, **KNN 알고리즘을 사용할 경우 생길 수 있는 문제점**으로 가장 적절한 것은?

1. 연봉의 영향이 너무 작아져 예측 정확도가 떨어진다  
2. 키의 영향이 너무 커져 예측 정확도가 떨어진다  
3. 연봉의 값이 너무 커서 거리 계산에서 **지나치게 큰 영향**을 미친다  
4. 특성 수가 2개이므로 문제 없음  

👉 정답: ( 3 )  
👉 간단한 이유: KNN 알고리즘은 특성 간 거리를 기준으로 예측을 수행한다. 그런데 두 특성의 값 범위가 너무 다를 경우, 값이 큰 특성(예: 연봉) 이 거리 계산에서 절대적으로 우세한 영향을 미치게 된된다. 
    이로 인해 상대적으로 중요할 수 있는 작은 특성(예: 키) 의 영향은 거의 반영되지 않고, 모델이 왜곡된 거리 기반 판단을 하게 되어 예측 성능이 떨어질 수 있다.

## 문제 8. train_test_split() 함수의 목적
다음 중 `train_test_split()` 함수를 사용하는 **가장 주요한 목적**은 무엇인가?

1. 데이터를 시각화하기 위해 데이터를 나누는 함수  
2. 전체 데이터를 더 빠르게 처리하기 위해 나누는 함수  
3. 과적합을 방지하기 위해 학습용 데이터와 테스트용 데이터를 구분하는 함수  
4. 모델의 정확도를 높이기 위해 데이터를 랜덤하게 삭제하는 함수  

👉 정답: ( 3 )  
👉 간단한 이유: 만약 모델 튜닝에 모든 데이터를 사용하게 될 경우, 해당 데이터에 맞게 튜닝이 되어 과적합이 될 수 있어서 이에 따른 일반화 성능이 떨어질 수 있다.

fitness_user_data.csv 파일을 pandas를 사용하여 로드하세요.

데이터의 기본 정보(info()), 처음 5개 행(head()), 결측값 현황(isnull().sum())을 확인하세요.

2. 데이터 전처리
   
(1) 결측값 처리

**exercise_time (운동 시간, 분)**의 결측값은 평균값으로 채우세요.

**calories_burned (소모 칼로리)**의 결측값은 0으로 채우세요. (운동을 기록하지 않은 경우도 있을 수 있음)

**weight (몸무게, kg)**의 결측값은 최빈값(가장 많이 등장한 값)으로 채우세요.

(2) 이상치 처리
exercise_time 중 180분 이상 운동한 데이터는 이상치로 간주하고, 90% 백분위수 값으로 대체하세요.

calories_burned 중 1500 이상인 값은 이상치로 간주하고, 75% 백분위수 값으로 대체하세요.

(3) 날짜 처리
exercise_date 열이 문자열 형식으로 되어 있다면, 이를 datetime 형식으로 변환하세요.

**각 운동 날짜의 요일(day_of_week)**을 추출하여 **새로운 열 exercise_day_of_week**을 추가하세요.

3. 데이터 시각화 (산점도 활용)

운동 시간과 소모 칼로리 관계 (산점도)

X축: exercise_time (운동 시간)

Y축: calories_burned (소모 칼로리)

포인트로 개별 사용자를 시각화하세요.

운동 시간이 많을수록 더 많은 칼로리를 소모하는지 확인하세요.

운동 유형과 소모 칼로리 관계 (산점도)

X축: exercise_type (운동 유형, 문자열이므로 변환 필요)

Y축: calories_burned (소모 칼로리)

포인트로 개별 사용자를 시각화하세요.

특정 운동 유형(예: 러닝 vs. 요가 vs. 웨이트 트레이닝)이 더 많은 칼로리를 소모하는지 확인하세요.

운동 시간과 몸무게 관계 (산점도)

X축: exercise_time (운동 시간)

Y축: weight (몸무게)

운동 시간이 많을수록 몸무게가 줄어드는지 분석하세요.

특정 몸무게 구간에서 운동 시간이 길어지는 경향이 있는지 확인하세요.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. 데이터 로드 및 기본 탐색
df = pd.read_csv('fitness_user_data.csv')
print("데이터 정보:")
print(df.info())
print("\n결측값 개수:")
print(df.isnull().sum())
df.head()
df
```

```python

# 2. 결측값 처리
df['exercise_time'] = df['exercise_time'].fillna(df['exercise_time'].mean())
df['calories_burned'] = df['calories_burned'].fillna(0)
df['weight'] = df['weight'].fillna(df['weight'].mode()[0])
df
```

```python
# 3. 이상치 처리
q90 = df['exercise_time'].quantile(0.90)  # 90% 백분위수 값
print('q90=', q90)
df['exercise_time'] = df['exercise_time'].apply(lambda x: q90 if x >= 180 else x)

q3 = df['calories_burned'].quantile(0.75)  # 75% 백분위수 값
print('q3=', q3)
df['calories_burned'] = df['calories_burned'].apply(lambda x: q3 if x >= 1500 else x)


df = df[df['exercise_time'] > 0]  # 0 이하인 운동 시간 제거
df
```

```python
# 4. 날짜 처리
df['exercise_date'] = pd.to_datetime(df['exercise_date'])
df['exercise_day_of_week'] = df['exercise_date'].dt.day_name() # .dt: This accessor provides access to datetime-related properties and methods.
df
```

```python
# 5. 데이터 시각화 (산점도 사용)

# (1) 운동 시간과 소모 칼로리 관계 (산점도)
plt.figure(figsize=(8,5))
plt.scatter(df['exercise_time'], df['calories_burned'], color='blue', alpha=0.5)
plt.xlabel('Exercise Time (minutes)')
plt.ylabel('Calories Burned')
plt.title('Exercise Time vs. Calories Burned')
plt.grid(True)
plt.show()
```

```python
# (2) 운동 유형과 소모 칼로리 관계 (산점도)
exercise_mapping = {'Running': 1, 'Weight Training': 2, 'Yoga': 3}
df['exercise_type_numeric'] = df['exercise_type'].map(exercise_mapping)
df
```

```python

plt.figure(figsize=(8,5))
plt.scatter(df['exercise_type_numeric'], df['calories_burned'], color='green', alpha=0.5)
plt.xlabel('Exercise Type (1: Running, 2: Weight Training, 3: Yoga)')
plt.ylabel('Calories Burned')
plt.title('Exercise Type vs. Calories Burned')
plt.xticks([1, 2, 3], ['Running', 'Weight Training', 'Yoga'])
plt.grid(True)
plt.show()
```

```python
# (3) 운동 시간과 몸무게 관계 (산점도)
plt.figure(figsize=(8,5))
plt.scatter(df['exercise_time'], df['weight'], color='red', alpha=0.5)
plt.xlabel('Exercise Time (minutes)')
plt.ylabel('Weight (kg)')
plt.title('Exercise Time vs. Weight')
plt.grid(True)
plt.show()
```

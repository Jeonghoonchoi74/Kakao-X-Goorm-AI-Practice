# 영화 데이터 분석 문제
https://pandas.pydata.org/ 사이트의 매뉴얼을 참조해서 
강의 시간에 다루지 않았던 pandas 패키지의 다양한 메소드 함수를 찾아서 사용법을 파악해서 풀어 보는 문제입니다.

설명:
15개의 영화 데이터가 포함된 데이터프레임이 주어진다.

영화의 **제목, 장르, 개봉 연도, 감독, 평점, 관객 수(단위: 백만명), 제작비(단위: 백만 달러)**가 포함된다.

데이터프레임에서 직접 인덱싱과 슬라이싱 등을 활용하여 문제를 해결하세요.

```python
import pandas as pd

data = {
    "Title": [
        "Inception", "Interstellar", "The Dark Knight", "Parasite", "Avengers: Endgame",
        "The Matrix", "Gladiator", "Titanic", "Joker", "Mad Max: Fury Road",
        "The Godfather", "Pulp Fiction", "Schindler's List", "The Shawshank Redemption", "Fight Club"
    ],
    "Genre": [
        "Sci-Fi", "Sci-Fi", "Action", "Drama", "Action",
        "Sci-Fi", "Historical", "Romance", "Drama", "Action",
        "Crime", "Crime", "Drama", "Drama", "Drama"
    ],
    "Year": [
        2010, 2014, 2008, 2019, 2019,
        1999, 2000, 1997, 2019, 2015,
        1972, 1994, 1993, 1994, 1999
    ],
    "Director": [
        "Christopher Nolan", "Christopher Nolan", "Christopher Nolan", "Bong Joon-ho", "Anthony Russo",
        "Lana Wachowski", "Ridley Scott", "James Cameron", "Todd Phillips", "George Miller",
        "Francis Ford Coppola", "Quentin Tarantino", "Steven Spielberg", "Frank Darabont", "David Fincher"
    ],
    "Rating": [
        8.8, 8.6, 9.0, 8.6, 8.4,
        8.7, 8.5, 7.8, 8.5, 8.1,
        9.2, 8.9, 8.9, 9.3, 8.8
    ],
    "Audience_Million": [
        100, 120, 150, 80, 200,
        90, 50, 220, 85, 75,
        110, 95, 85, 130, 70
    ],
    "Budget_Million": [
        160, 165, 185, 11, 356,
        63, 103, 200, 55, 150,
        6, 8, 22, 25, 63
    ]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
"""
1. 특정 장르의 영화 목록 출력
장르가 "Action"인 영화만 출력하세요.
📌 힌트: df["Genre"] == "Action"을 활용하여 필터링."
"""
genre = df[df["Genre"] == "Action"]

genre
```

```python
"""
2. 특정 연도 이후 개봉한 영화 목록
2010년 이후 개봉한 영화만 출력하세요.
"""
df[df["Year"] >= 2010]
```

```python
"""
3. 평점이 9.0 이상인 영화 찾기
평점이 9.0 이상인 영화만 출력하세요.
"""
df[df["Rating"]>=9.0]
```

```python
"""
4. 특정 감독의 영화 목록 출력
"Christopher Nolan"이 감독한 영화만 출력하세요.
"""
df[df["Director"] == "Christopher Nolan" ]
```

```python
"""
5. 특정 예산 범위 내의 영화 찾기
제작비가 50백만 ~ 200백만 사이인 영화만 출력하세요.
"""
df[(df["Budget_Million"] > 50) & (df["Budget_Million"] < 200)]
```

```python
"""
6. 제작비 100백만 이상인 영화만 필터링
제작비가 100백만 이상인 영화 목록 출력
"""
df[df["Budget_Million"]>=100]
```

```python
"""
7. 관객 수가 100백만 이상인 영화만 찾기
관객 수가 100백만 이상인 영화 출력
"""

df[df["Audience_Million"]>=100]
```

```python
"""
8. 가장 최근에 개봉한 영화 찾기
연도가 가장 큰 값에 해당하는 영화 찾기 (max() 사용 금지)
📌 힌트: .iloc[]을 활용하여 정렬 후 마지막 행 선택.
"""
year = df.sort_values(by="Year", ascending=False)
max_year = year.iloc[0]["Year"]
max_year
df[df["Year"] == max_year]
```

```python
"""
9. 가장 오래된 영화 찾기
연도가 가장 작은 값에 해당하는 영화 찾기 (min() 사용 금지)
"""
year = df.sort_values(by="Year", ascending=True)
min_year = year.iloc[0]["Year"]
min_year
df[df["Year"] == min_year]
```

```python
"""
10. 첫 5개의 영화만 출력
데이터프레임의 첫 5개 행만 출력하세요.
📌 힌트: df[:5] 또는 df.head(5) 없이 해결.
"""
sliced = df.iloc[:5]
sliced


```

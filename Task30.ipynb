## âœ… ë¬¸ì œ 2. ì´ì§„ ê²€ìƒ‰ êµ¬í˜„
**ì„¤ëª…**: ì •ë ¬ëœ ë°°ì—´ì—ì„œ ì´ì§„ ê²€ìƒ‰(binary search)ì„ êµ¬í˜„í•˜ì„¸ìš”.  
- ì…ë ¥: ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬ëœ ì •ìˆ˜ ë°°ì—´ `arr`ê³¼ ì •ìˆ˜ `target`
- ì¶œë ¥: `target`ì˜ ì¸ë±ìŠ¤, ì—†ìœ¼ë©´ `-1`
```python
def binary_search(arr: list[int], target: int) -> int:
    pass  # ì—¬ê¸°ì— êµ¬í˜„
```
**ì˜ˆì‹œ**
```python
>>> binary_search([1, 3, 5, 7, 9], 7)
3
>>> binary_search([1, 3, 5, 7, 9], 4)
-1
```

````python
# binary_search.py

def binary_search(arr: list[int], target: int) -> int:
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


target = int(input("ì°¾ê³ ì í•˜ëŠ” ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
arr2 = list(map(int, input("ì •ìˆ˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ").split()))

arr2.sort()  # ì´ì§„ íƒìƒ‰ì€ ì •ë ¬ì´ í•„ìš”

print("ì…ë ¥ëœ ë¦¬ìŠ¤íŠ¸:", arr2)
print("ì°¾ê³ ì í•˜ëŠ” ìˆ«ì:", target)

result = binary_search(arr2, target)
if result != -1:
    print(f"ìˆ«ì {target}ì€(ëŠ”) ì¸ë±ìŠ¤ {result}ì— ìˆìŠµë‹ˆë‹¤.")
else:
    print(f"ìˆ«ì {target}ì€(ëŠ”) ë¦¬ìŠ¤íŠ¸ì— ì—†ìŠµë‹ˆë‹¤.")

````

## âœ… ë¬¸ì œ 3. ì²´ì´ë‹ í•´ì‹œë§µ êµ¬í˜„ (ê¸°ì´ˆ ë²„ì „)
**ì„¤ëª…**: í•´ì‹œ ì¶©ëŒì„ ì²´ì´ë‹(chaining) ë°©ì‹ìœ¼ë¡œ í•´ê²°í•˜ëŠ” í•´ì‹œë§µì„ í´ë˜ìŠ¤ í˜•íƒœë¡œ êµ¬í˜„í•˜ì„¸ìš”.
- ì œê³µ ê¸°ëŠ¥:
  - `put(key, value)`: í‚¤-ê°’ ì‚½ì…
  - `get(key)`: í‚¤ë¡œ ê°’ ê²€ìƒ‰

```python
class ChainedHashMap:
    def __init__(self, size: int = 10):
        pass
    def put(self, key: str, value: int) -> None:
```

````python
# chaining hashmap

class ChainedHashMap:
    def __init__(self, size: int = 10):
        self.size = size
        self.table = [[] for _ in range(size)]
    
    def hash(self, key: str) -> int:
        # ë¬¸ìì—´ keyë¥¼ ì •ìˆ˜ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ëŠ” ê°„ë‹¨í•œ í•´ì‹œ í•¨ìˆ˜
        return sum(ord(c) for c in key) % self.size

    def put(self, key: str, value: int) -> None:
        # íƒìƒ‰í•œ ë’¤ keyê°€ ì¡´ì¬í•˜ë©´ valueë¥¼ ì—…ë°ì´íŠ¸
        for i, (k, v) in enumerate(self.table[self.hash(key)]):
            if k == key:
                self.table[self.hash(key)][i] = (key, value)
                return
        # keyê°€ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ ìƒˆë¡œìš´ ì—”íŠ¸ë¦¬ë¥¼ ì¶”ê°€
        self.table[self.hash(key)].append((key, value))

    def get(self, key: str) -> int | None:
        # íƒìƒ‰í•œ ë’¤ keyê°€ ì¡´ì¬í•˜ë©´ valueë¥¼ ë°˜í™˜
        for k, v in self.table[self.hash(key)]:
            if k == key:
                return v
        # keyê°€ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ Noneì„ ë°˜í™˜
        return -1
    
    def remove(self, key: str) -> None:
        # íƒìƒ‰í•œ ë’¤ keyê°€ ì¡´ì¬í•˜ë©´ ì‚­ì œ
        for i, (k, v) in enumerate(self.table[self.hash(key)]):
            if k == key:
                del self.table[self.hash(key)][i]
                return
        # keyê°€ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ -1 ë°˜í™˜
        return -1



hashmap = ChainedHashMap(size=10)

print("Chained HashMap")
print("---------------")
print("1. ì¶”ê°€")
print("2. ì¡°íšŒ")
print("3. ì‚­ì œ")
print("4. ì¢…ë£Œ")
print("---------------")

while True:
    choice = int(input("ì›í•˜ëŠ” ì‘ì—…ì„ ì„ íƒí•˜ì„¸ìš”: "))

    if choice == 1:
        print("ì¶”ê°€")
        key = input("í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        value = int(input("ê°’ì„ ì…ë ¥í•˜ì„¸ìš”: "))
        hashmap.put(key, value)
        print(f"{key} : {value} ì¶”ê°€ ì™„ë£Œ")

    elif choice == 2:
        print("ì¡°íšŒ")
        key = input("ì¡°íšŒí•  í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        value = hashmap.get(key)
        if value != -1:
            print(f"{key} : {value}")
        else:
            print(f"{key}ëŠ”(ì€) ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

    elif choice == 3:
        print("ì‚­ì œ")
        key = input("ì‚­ì œí•  í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        value = hashmap.remove(key)
        if value != -1:
            print(f"{key} ì‚­ì œ ì™„ë£Œ")
        else:
            print(f"{key}ëŠ”(ì€) ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

    elif choice == 4:
        print("ì¢…ë£Œ.")
        break

    else:
        print("ì˜ëª»ëœ ì„ íƒ.")
````

## âœ… ë¬¸ì œ 4. í•´ì‹œ ì¶©ëŒ ì‹œë‚˜ë¦¬ì˜¤ ì‹¤ìŠµ
**ì„¤ëª…**: ê°™ì€ í•´ì‹œê°’ì´ ë‚˜ì˜¤ë„ë¡ `size=1`ì¸ í•´ì‹œë§µì„ ë§Œë“¤ê³ , ì—¬ëŸ¬ keyë¥¼ ë„£ê³  ì²´ì´ë‹ ì¶©ëŒ ìƒí™©ì„ ê´€ì°°í•˜ì„¸ìš”.

```python
# ì§ì ‘ put ì—¬ëŸ¬ ë²ˆ ì‹¤í–‰ í›„ ë‚´ë¶€ êµ¬ì¡°ë¥¼ ì¶œë ¥í•´ë³´ì„¸ìš”.
```

````python
# hash collision, set hash size 1

class ChainedHashMap:
    def __init__(self, size: int = 1):
        self.size = size
        self.table = [[] for _ in range(size)]
    
    def hash(self, key: str) -> int:
        return sum(ord(c) for c in key) % self.size

    def put(self, key: str, value: int) -> None:
        # íƒìƒ‰í•œ ë’¤ keyê°€ ì¡´ì¬í•˜ë©´ valueë¥¼ ì—…ë°ì´íŠ¸
        for i, (k, v) in enumerate(self.table[self.hash(key)]):
            if k == key:
                self.table[self.hash(key)][i] = (key, value)
                return
        # keyê°€ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ ìƒˆë¡œìš´ ì—”íŠ¸ë¦¬ë¥¼ ì¶”ê°€
        self.table[self.hash(key)].append((key, value))

    def get(self, key: str) -> int | None:
        # íƒìƒ‰í•œ ë’¤ keyê°€ ì¡´ì¬í•˜ë©´ valueë¥¼ ë°˜í™˜
        for k, v in self.table[self.hash(key)]:
            if k == key:
                return v
        # keyê°€ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ Noneì„ ë°˜í™˜
        return -1
    
    def remove(self, key: str) -> None:
        # íƒìƒ‰í•œ ë’¤ keyê°€ ì¡´ì¬í•˜ë©´ ì‚­ì œ
        for i, (k, v) in enumerate(self.table[self.hash(key)]):
            if k == key:
                del self.table[self.hash(key)][i]
                return
        # keyê°€ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ -1 ë°˜í™˜
        return -1
    

hashmap = ChainedHashMap(size=1)

print("Chained HashMap")
print("---------------")
print("1. ì¶”ê°€")
print("2. ì¡°íšŒ")
print("3. ì‚­ì œ")
print("4. ì¢…ë£Œ")
print("5. ì „ì²´ ì¡°íšŒ")
print("---------------")

while True:
    choice = int(input("ì›í•˜ëŠ” ì‘ì—…ì„ ì„ íƒí•˜ì„¸ìš”: "))

    if choice == 1:
        print("ì¶”ê°€")
        key = input("í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        value = int(input("ê°’ì„ ì…ë ¥í•˜ì„¸ìš”: "))
        hashmap.put(key, value)
        print(f"{key} : {value} ì¶”ê°€ ì™„ë£Œ")

    elif choice == 2:
        print("ì¡°íšŒ")
        key = input("ì¡°íšŒí•  í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        value = hashmap.get(key)
        if value != -1:
            print(f"{key} : {value}")
        else:
            print(f"{key}ëŠ”(ì€) ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

    elif choice == 3:
        print("ì‚­ì œ")
        key = input("ì‚­ì œí•  í‚¤ë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")
        value = hashmap.remove(key)
        if value != -1:
            print(f"{key} ì‚­ì œ ì™„ë£Œ")
        else:
            print(f"{key}ëŠ”(ì€) ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

    elif choice == 4:
        print("ì¢…ë£Œ.")
        break

    # ì „ì²´ ì¡°íšŒ
    elif choice == 5:
        print("ì „ì²´ ì¡°íšŒ")
        for i, bucket in enumerate(hashmap.table):
            print(f"ë²„í‚· {i}: {bucket}")
    else:
        print("ì˜ëª»ëœ ì„ íƒ.")
````

## âœ… ë¬¸ì œ 5. ê²€ìƒ‰ ì•Œê³ ë¦¬ì¦˜ ë¹„êµ ì‹¤í—˜

````python
# time taken for each method linear, binary

import time
import random

def linear_search(arr: list[int], target: int) -> int:
    for index, value in enumerate(arr):
        if value == target:
            return index
    return -1

def binary_search(arr: list[int], target: int) -> int:
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


random_list = list(range(100000000))  # 1,000,000ê°œì˜ ì •ìˆ˜ë¡œ ì´ë£¨ì–´ì§„ ë¦¬ìŠ¤íŠ¸
target = random.choice(random_list)  # ë¦¬ìŠ¤íŠ¸ì—ì„œ ë¬´ì‘ìœ„ë¡œ ìˆ«ì ì„ íƒ

print("ë¬´ì‘ìœ„ ë¦¬ìŠ¤íŠ¸:", random_list[:10], "...")  # ì²˜ìŒ 10ê°œë§Œ ì¶œë ¥
print("ì°¾ê³ ì í•˜ëŠ” ìˆ«ì:", target)

# ì„ í˜• íƒìƒ‰
start_time = time.time()
result = linear_search(random_list, target)
end_time = time.time()
print(f"ì„ í˜• íƒìƒ‰ì˜ ì¸ë±ìŠ¤ ìœ„ì¹˜: {result}")
print(f"ì„ í˜• íƒìƒ‰ ì‹œê°„: {end_time - start_time:.6f}ì´ˆ")

# ì´ì§„ íƒìƒ‰

start_time = time.time()
result = binary_search(random_list, target)
end_time = time.time()
print(f"ì´ì§„ íƒìƒ‰ì˜ ì¸ë±ìŠ¤ ìœ„ì¹˜: {result}")
print(f"ì´ì§„ íƒìƒ‰ ì‹œê°„: {end_time - start_time:.6f}ì´ˆ")
````
